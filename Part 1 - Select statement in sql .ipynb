# How to use select statement ?

```python
import pandas as pd
import pandasql as ps
```

```python
ufo_cols = ['city', 'colors_reported', 'shape_reported', 'state', 'time']
ufo=pd.read_csv('http://bit.ly/uforeports',names = ufo_cols,header = 0)
```

header = 0 indicates the data rows should start from the first row(index 0) with city "Ithaca" and not from the heading which is indexed as -1, by default the output of the code without mentioned header is same as that of header = -1

```python
ufo.head()
```

```python
a1 = "SELECT * FROM ufo LIMIT 10"
```

```python
ps.sqldf(a1, locals())
```

The select statement retrives data from a table,
the * indicates to retrive all the rows & columns without filtering
and the limit keyword limits the fetched data to the number specified with it, in this case it is 10

```python
b1 = """SELECT DISTINCT shape_reported  FROM ufo """
```

```python
ps.sqldf(b1, locals())
```

The DISTINCT keyword only fetches non duplicate values for the column name next to it
For Eg - if there are 6 entries 3 entries with City "Alma" 2 with "Hubbard" and 1 entry with city "Crater Lake"
In such a case the output will be of only 3 rows with the 3 city names - (Alma, Hubbard & Crater Lake) excluding the multiple entries for each city names.
